## Creamos las tablas que usaremos en la base de datos

### Creamos la tabla de los datos de los empleados

```python
# Ahora vamos a leer el archivo CSV de empleados
df = pd.read_csv('empleados.csv')
df
```

```python
df.dtypes
```

```python
# Cambiaremos varios tipos de datos para que se ajusten a la base de datos
df['id'] = df['id'].astype('int64')
df['nombre'] = df['nombre'].astype('string')
df['finca'] = df['finca'].astype('string')
df['cedula'] = df['cedula'].astype('string')
df['labor_usual'] = df['labor_usual'].astype('string')
df['estado'] = df['estado'].astype('string')
df['fecha_ingreso'] = pd.to_datetime(df['fecha_ingreso'], errors='coerce')
df['fecha_contrato'] = pd.to_datetime(df['fecha_contrato'], errors='coerce')
df['fecha_vencimiento'] = pd.to_datetime(df['fecha_vencimiento'], errors='coerce')
df['tipo'] = df['tipo'].astype('string')
df['estado'] = df['estado'].astype('string')
df['telefono'] = df['telefono'].astype('string')
df['cuenta_bancaria'] = df['cuenta_bancaria'].astype('string')
df['tipo_cuenta'] = df['tipo_cuenta'].astype('string')
df['banco'] = df['banco'].astype('string')
df['nombre_cuenta'] = df['nombre_cuenta'].astype('string')
df.dtypes
```

```python
df
```

```python
# Usamos la columna id como índice
df.set_index('id', inplace=True)
df
```

```python
# Ahora vamos a crear la tabla en la base de datos, asignando la columna id como llave primaria y los valores de las columnas 'fecha_ingreso', 'fecha_contrato' y 'fecha_vencimiento' como fechas
database_connection.connect()
df.to_sql('empleados', database_connection, if_exists='replace', index=True, index_label='id', dtype={'fecha_ingreso': db.DATE, 'fecha_contrato': db.DATE, 'fecha_vencimiento': db.DATE})
print("Tabla empleados creada exitosamente")
```
